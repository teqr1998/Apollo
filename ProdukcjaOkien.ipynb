```
problem = LpProblem("Max_zysk", LpMaximize)
okna_a = LpVariable("Aluminiowe")
okna_d = LpVariable("Drewniane")
problem += okna_a * 3 + okna_d * 5, "Zysk"
problem += okna_a * 1 <= 4, "Ogr_1"
problem += okna_d * 2 <= 12, "Ogr_2"
problem += okna_a * 3 + okna_d * 2 <= 18, "Ogr_3"
problem.solve()
print("Największy zysk: " + str(problem.objective.value()) + " tys.")
for v in problem.variables():
    print(v.varValue, v.name)
```

```
from pulp import *
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd
```

```
zysk_okna_d = FloatSlider(min = 0, max = 100, step = 0.5, value = 5, description =  "Zysk drewniane")
zysk_okna_a = FloatSlider(min = 0, max = 100,  step = 0.5, value = 3,  description = "Zysk aluminiowe")
ogr_1 = IntSlider(min = 0, max = 100, value = 4,  description = "Ogr. 1")
ogr_2 = IntSlider(min = 0, max = 100, value = 12, description = "Ogr. 2")
ogr_3 = IntSlider(min = 0, max = 100, value = 18, description = "Ogr. 3")

def max_zysk(zysk_d, zysk_a, limit_1, limit_2, limit_3):
    problem = LpProblem("Maksymalizacja_zyskow_okien", LpMaximize)
    okna_a = LpVariable("Aluminiowe", lowBound=0)
    okna_d = LpVariable("Drewniane", lowBound=0)
    problem += okna_a * zysk_a + okna_d * zysk_d, "Liczb_partii_wszystkich_okien"
    problem += okna_a * 1 <= limit_1, "Ogr_1"
    problem += okna_d * 2 <= limit_2, "Ogr_2"
    problem += okna_a * 3 + okna_d * 2 <= limit_3, "Ogr_3"
    problem.solve()
    print("Zysk: " + str(problem.objective.value()))
    for v in problem.variables():
        print(v.name +" "+ str(v.varValue))
    for k, v in problem.constraints.items():
        print(k + " slack: " + str(v.slack) + " shadow price: " + str(v.pi))

interact(max_zysk, zysk_d=zysk_okna_d, zysk_a=zysk_okna_a, limit_1=ogr_1, limit_2=ogr_2, limit_3=ogr_3)
```
