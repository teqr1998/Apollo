```
dni = { "pn": 2, "wt": 1, "śr": 1, "cz": 1, "pt": 3 }
pracownicy = [{ "Ania": 1,"Stefan": 1,"Hektor": 1,"Olaf": 1,"Lidia": 0, "Piotr": 1 },
{ "Ania": 1,"Stefan": 0,"Hektor": 1,"Olaf": 1,"Lidia": 0, "Piotr": 1 },
{ "Ania": 1,"Stefan": 0,"Hektor": 1,"Olaf": 1,"Lidia": 0, "Piotr": 1 },
{ "Ania": 1,"Stefan": 0,"Hektor": 0,"Olaf": 1,"Lidia": 1, "Piotr": 0 },
{ "Ania": 1,"Stefan": 0,"Hektor": 0,"Olaf": 1,"Lidia": 1, "Piotr": 0 }]
pensja = { "Ania": 100,"Stefan": 50,"Hektor": 60,"Olaf": 40,"Lidia": 110,"Piotr": 70 }
```

```
grafik = []
koszty = {}
for n, d in enumerate(dni.keys()):
    for prac in pensja.keys():
        grafik.append(d + "_" + prac)
        koszty[grafik[-1]] = pracownicy[n][prac] * pensja[prac]

lista = LpVariable.dicts("Grafik_", grafik, cat=LpBinary)

problem = LpProblem("Grafik_sklepowy", LpMinimize)
problem += lpSum([lista[x] * p for x, p in koszty.items()])
# Dostępność
for x, p in koszty.items():
    problem += lista[x] * p <= koszty[x]
# Obsługa w danym dniu
for d in dni.keys():
    problem += lpSum([lista[x] for x, p in  koszty.items() if d in x]) == dni[d]
# Tygodniowy czas pracy
for prac in pensja.keys():
    problem += lpSum([lista[x] for x, p in  koszty.items() if prac in x]) <= 3
# Rozwiązanie problemu
problem.solve()
for v in problem.variables():
    print(v.name +" "+ str(v.varValue))
```
